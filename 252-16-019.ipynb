```python
mark= int(input("Enter your mark: "))
if(mark>=90 and mark<=100):
      grade="A+"
elif(mark>=80 and mark<=89):
      grade="A"
elif(mark>=60 and mark<=79):
      grade="B"
elif(mark>=40 and mark<=59):
      grade="C"
else:
      grade="F"
if(grade=="F"):
   print("Fail")
else:
    print("Pass")
    print(f"Grade: {grade}")
```
